```python
#First we import all the libraries that we need to proceed to this task
import pandas as pd
import numpy as np 
import spacy
from sklearn.metrics import accuracy_score
from textblob import TextBlob
```

```python
#Introducing the dataframe
df= pd.read_csv('/kaggle/input/consumer-reviews-of-amazon-products/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv')
df.head()
```

```python
#Select the column we want to work with 'reviews.text' and cleaning it
clean= df[['reviews.text']]
clean.head()
```

```python
clean.isnull()
```

```python
#Indroducing nlp with spaCy 
text= clean['reviews.text']
text
nlp= spacy.load('en_core_web_sm')

#Tokenization and removing the stopwords
def preprocess(text):
    doc= nlp(text.lower().strip())
    processed= [token.lemma_ for token in doc if not token.is_stop]
    #print(processed)
    return ' '.join(processed)
```

```python
clean['processed_text']= clean['reviews.text'].apply(preprocess)
data= clean['processed_text']
data.head()
```

```python
#Set the conditions for the polarity and sentiments 
def analyze_polarity(reviews):
    polarity_scores= []
    for review in df['reviews.text']:
        
        doc=nlp(review)
        blob= TextBlob(review)
        polarity= blob.sentiment.polarity
        polarity_scores.append(polarity)
    return polarity_scores
```

```python
reviews = df['reviews.text']
polarity_scores = analyze_polarity(reviews)
 
sentiments= [ ] 
for score in polarity_scores:
    
    if score >0:
        sentiment= 'positive'
    
    elif score <0:
        sentiment='negative'
    
    else:
        sentiment ='neutral'
sentiments.append(sentiment)
  
print('The overall sentiment in the reviews is ',sentiments,'.')
```

```python
#Calculating the similarity between 2 reviews 

def get_vector(text):
    doc= nlp(text)
    return doc.vector

clean['vector'] = clean['reviews.text'].apply(get_vector)
clean['vector'].iloc[0].shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def similarity(input_reviews, n_reviews = 1):
    input_vector= get_vector(preprocess(input_reviews))
    input_vector= input_vector.reshape(1,-1)
    
    similarity_scores= cosine_similarity(input_vector, list(clean['vector']))
    similarity_scores_indices= similarity_scores[0].argsort()[-n_reviews :][::-1]
    return similarity_scores_indices


review_choice= clean['reviews.text'][3]
review_choice_indices= similarity(review_choice)
print('Review number 3 is the most similar with review number:',review_choice_indices,'.')
```
